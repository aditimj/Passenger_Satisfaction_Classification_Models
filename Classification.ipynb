**Discussion:** When doing comparison of different classifiers, it is important to use corrext evaluation metrics to understand the performance of the classifiers.
- **Confusion Matrix:** Confusion Matrix is a square matrix which would give the count for true negative, false positive, false negative, and true positive. 
    - TN: Person is dissatisfied and is classifed as dissatisfied. (Correctly classfied)
    - FP: Person is dissatisfied but is classified as satisfied.
    - FN: Person is satisfied but is classfied as dissatisfied.
    - TP: Person is satisfied and is classfied as satisfied. (Correctly classfied)
- **Accuracy:** Accuracy represents the number of correctly classified data instances over the total number of data instances.
    - (TN+TP) / (TN+FP+FN+TP)
- **Recall:** Recall should ideally be 1. Which would mean that there are 0 False Negative (FN).
    - TP / (TP + FN)
- **Precision:** Recall should ideally be 1. Which would mean that there are 0 False Positive (FP).
    - TP / (TP + FP)
- **F1 Score:** F1 Score takes both precision and recall into account and F1 score value becomes 1 only when Precesion and Recall are 1.
    - 2 * ((Precision * Recall)/(Precision + Recall))
- **Net Benefit:** Would provide cost information. This is calculated below. 
    - Benefit - Cost
- **Net Benefit Per person:** This would be calculated based on total test inputs and total cost
    - Net Benefit / Total test set size

```python
def calculate_cost(conf_matrix):
    cost_matrix = np.array([[82.5, -450], [-75, 0]])
    
    total_cost = np.sum(conf_matrix * cost_matrix)

    return total_cost
    print(total_cost)
```

**Discussion:**

I have performed KNN for delfaut values. 
Here, k = 5


"k" represents the number of nearest neighbors to consider when making predictions or classifications.
KNN identifies the k data points in the training dataset that are closest to the given point in terms of a chosen distance metric. It then assigns the class based on the majority class or the average value among those k neighbors.



```python
warnings.simplefilter(action='ignore', category=FutureWarning)

results_knn = pd.DataFrame(columns=['K', 'Accuracy Score', 'Recall Score', 'Precision Score', 'F1 Score', 'Net Benefit', 'Net Benefit (Per Person)'])

k_values = list(range(1, 22, 2))  

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train) 
    
    knn_y_pred = knn.predict(X_test)  
    
    
    conf_matrix = confusion_matrix(y_test, knn_y_pred)
    accuracy = accuracy_score(y_test, knn_y_pred)
    recall = recall_score(y_test, knn_y_pred)
    precision = precision_score(y_test, knn_y_pred)
    f1 = f1_score(y_test, knn_y_pred)
    knn_total_cost = calculate_cost(conf_matrix)
    knn_total_cost_person = knn_total_cost / testset

    
    results_knn = results_knn.append({'K': k, 'Accuracy Score': accuracy,
                              'Recall Score': recall, 'Precision Score': precision, 'F1 Score': f1,
                              'Net Benefit': knn_total_cost, 'Net Benefit (Per Person)':knn_total_cost_person }, ignore_index=True)

pd.set_option('display.max_colwidth', None)
print(results_knn.to_string(index=False))
```

**Discussion:** In the above section, I have evaluated several values of k parameter.
Choosing the right value for the parameter "k" is a crucial decision that can significantly affect the performance of your model.

Specifically, I chose to evaluate odd values for K this being a binary classification <satistisfied ; dissatisfied or neutral>

This is because odd values help avoid ties when determining the majority class. Odd "k" values help avoid situations where there is an equal number of neighbors from each class, which can lead to ambiguous predictions.

```python
n_estimators_values = [100, 150, 200]
max_depth_values = list(range(1, 22, 2))  

results_rf = pd.DataFrame(columns=['n_estimators', 'max_depth', 'Accuracy Score', 'Recall Score', 'Precision Score', 'F1 Score', 'Net Benefit', 'Net Benefit (Per Person)'])


for n_estimators in n_estimators_values:
    for max_depth in max_depth_values:
        rf_classifier = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
        rf_classifier.fit(rf_X_train, rf_y_train)

        rf_y_pred = rf_classifier.predict(rf_X_test)

        rf_conf_matrix_com = confusion_matrix(rf_y_test, rf_y_pred)
        rf_accuracy_com = accuracy_score(rf_y_test, rf_y_pred)
        rf_recall_com = recall_score(rf_y_test, rf_y_pred)
        rf_precision_com = precision_score(rf_y_test, rf_y_pred)
        rf_f1_com = f1_score(rf_y_test, rf_y_pred)
        rf_total_cost_com = calculate_cost(rf_conf_matrix_com)  
        rf_total_cost_pp_com = rf_total_cost_com / testset
        # Store the results in the DataFrame
        results_rf = results_rf.append({
                                        'n_estimators': n_estimators,
                                        'max_depth': max_depth,
                                        'Accuracy Score': rf_accuracy_com,
                                        'Recall Score': rf_recall_com,
                                        'Precision Score': rf_precision_com,
                                        'F1 Score': rf_f1_com,
                                        'Net Benefit': rf_total_cost_com,
                                        'Net Benefit (Per Person)' :rf_total_cost_pp_com }, ignore_index=True)
```
